# Second Day Confirmation

Assume you're given tables with information about TikTok user sign-ups & confirmations through email & text. New users on TikTok sign up using their email address, & upon sign-up, each user receives a text message confirmation to activation their account.

Write a query to display the user IDs of those who did not confirm their sign-up on the first day, but confirmed on the second day.

Definition:

* `action_date` refers to the date when users activated their accounts & confirmed their sign-up through text messages

# Answer

This is the `emails` table.

```
CREATE TABLE emails (
	email_id smallint,
	user_id smallint,
	signup_date timestamp
);

INSERT INTO emails
VALUES (125, 7771, '06/14/2022 00:00:00'),
	   (433, 1052, '07/09/2022 00:00:00');

SELECT * FROM emails;
```

This is the `texts` table.

```
CREATE TABLE texts (
	text_id smallint,
	email_id smallint,
	signup_action text,
	action_date timestamp
);

INSERT INTO texts
VALUES (6878, 125, 'Confirmed', '06/14/2022 00:00:00'),
	   (6997, 433, 'Not Confirmed', '07/09/2022 00:00:00'),
	   (7000, 433, 'Confirmed', '07/10/2022 00:00:00');

SELECT * FROM texts;
```

<img src = "texts Table.png" width = "600" style = "margin:auto"/>

We'll have to join the tables & then compare the `signup_date` with the `action_date` column.

```
SELECT *
FROM emails
LEFT JOIN texts 
	ON emails.email_id = texts.email_id
WHERE signup_date + '1 day'::interval = action_date;
```

Only user 1052 confirmed his/her sign-up on the second day.
